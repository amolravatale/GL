### Monthly totals of international airline passengers, 1949 to 1960. Represented in 1000s. 

# Question: Split the data into train and test. Build ARIMA model on train data set and test it on test data set

# Question: Analyze performance

```python
import pandas as pd 
import numpy as np 
from sklearn.metrics import mean_squared_error
from math import sqrt
import itertools
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import statsmodels.api as sm
import statsmodels.tsa.api as smt
import statsmodels.formula.api as smf
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.stattools import adfuller
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Importing data
df = pd.read_csv('airline-passengers.csv')
```

```python
df.head()
```

```python
df.Timestamp = pd.to_datetime(df.Month,format='%Y-%m') 
df.index = df.Timestamp 
```

```python
df.head()
```

```python
#Creating train and test set 
train=df[0:int(len(df)*0.7)] 
test=df[int(len(df)*0.7):]
```

```python
#Plot data
train.Passengers.plot(figsize=(15,8), title= 'Monthly Airpassengers', fontsize=14)
test.Passengers.plot(figsize=(15,8), title= 'Monthly Airpassengers', fontsize=14)
```

```python
passenger_ts = train.Passengers
```

```python
decomposition = sm.tsa.seasonal_decompose(passenger_ts, model='multiplicative')
```

```python
fig = decomposition.plot()
fig.set_figwidth(8)
fig.set_figheight(6)
fig.suptitle('Decomposition of multiplicative time series')
plt.show()
```

### ADF Test for Stationarity. Null hypothesis is underlying data is non-stationary

```python
dftest = adfuller(passenger_ts)
dftest
```

```python
fig, axes = plt.subplots(1, 2)
fig.set_figwidth(12)
fig.set_figheight(4)
smt.graphics.plot_acf(passenger_ts, lags=18, ax=axes[0])
smt.graphics.plot_pacf(passenger_ts, lags=18, ax=axes[1])
plt.tight_layout()
```

```python
passenger_ts_diff = passenger_ts.diff(periods=1)
passenger_ts_diff.dropna(inplace=True)
```

```python
fig, axes = plt.subplots(1, 2)
fig.set_figwidth(12)
fig.set_figheight(4)
smt.graphics.plot_acf(passenger_ts_diff, lags=12, ax=axes[0])
smt.graphics.plot_pacf(passenger_ts_diff, lags=12, ax=axes[1])
plt.tight_layout()
```

```python
dftest = adfuller(passenger_ts_diff)
dftest
```

Trend Differencing

```python
dftest = adfuller(passenger_ts_12_1)
dftest
```

```python
fig, axes = plt.subplots(1, 2)
fig.set_figwidth(12)
fig.set_figheight(4)
smt.graphics.plot_acf(passenger_ts_12_1, lags=18, ax=axes[0])
smt.graphics.plot_pacf(passenger_ts_12_1, lags=18, ax=axes[1])
plt.tight_layout()
```

```python
trendArima_pdq = (1,1,1)  ## (p, d, q)
```

```python
seasonalArima_pdqm = (0,1,0,12)   ### (P, D, Q)_m  (0,1,0,12)
```

```python
modelSarima = sm.tsa.statespace.SARIMAX(passenger_ts,
                                        order = trendArima_pdq,
                                        seasonal_order = seasonalArima_pdqm)
```

```python
trainedModel = modelSarima.fit()
```

```python
pred99 = trainedModel.get_forecast(steps=44, alpha=0.1)
```

```python
testResults = pd.DataFrame()
testResults['Passengers'] = test.Passengers
testResults['pass_forecasted'] = pd.Series(pred99.predicted_mean)
```

```python
#Plot data
train.Passengers.plot()
testResults.Passengers.plot()
testResults.pass_forecasted.plot()
```

```python
testResults['pae'] = (testResults.pass_forecasted - testResults.Passengers)
testResults['pae'] = np.abs((testResults.pae)/testResults.Passengers*100)
```

```python
np.mean(testResults.pae)
```

### Model checking

```python
trainedModel.plot_diagnostics(lags=18, figsize=(16,12))
plt.show()
```

### Optimizing ARIMA model

```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)
# Generate all different combinations of p, d and q triplets
pdq = list(itertools.product(p, d, q))
```

```python
pdq
```

```python
# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
seasonal_pdq
```

```python
best_aic = np.inf
best_pdq = None
best_seasonal_pdq = None
temp_model = None
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        
        try:
            temp_model = sm.tsa.statespace.SARIMAX(passenger_ts,
                                             order = param,
                                             seasonal_order = param_seasonal)
            results = temp_model.fit()
            if results.aic < best_aic:
                best_aic = results.aic
                best_pdq = param
                best_seasonal_pdq = param_seasonal
        except:
            continue
print("Best SARIMAX{}x{}0 model - AIC:{}".format(best_pdq, best_seasonal_pdq, best_aic))
```

```python
best_model = sm.tsa.statespace.SARIMAX(passenger_ts,
                                      order=(1, 1, 0),
                                      seasonal_order=(1, 1, 0, 12))
best_results = best_model.fit()
```

```python
bestPred99 = best_results.get_forecast(steps=44, alpha=0.1)
testResults['bestPass_forecasted'] = pd.Series(bestPred99.predicted_mean)
```

```python
testResults['bestPae'] = (testResults.bestPass_forecasted - testResults.Passengers)
testResults['bestPae'] = np.abs((testResults.bestPae)/testResults.Passengers*100)
```

```python
np.mean(testResults.bestPae)
```
