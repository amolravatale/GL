# Plots to explore variation

## Histograms

Divide data equally into bins (hyperparameter) and count how many observations are in each bin

## Density plots

Smoothen out the histogram and correct for dependency on the number of bins. Very useful to quickly infer the number of peaks in the distribution

## CDF plots

Present a view of the fraction of observations that fall below specific values in the sample. This plot has no hyperparameters. Can be very useful to identify skew

## Boxplots

Summarize the quartiles and outliers in one visualization

```python
plt.figure(figsize=(15, 5))
sns.histplot(nestle_data.Close, 
             bins=10,
             kde=True)
```
