# K nearest neighbors

KNN falls in the supervised learning family of algorithms. Informally, this means that we are given a labelled dataset consiting of training observations (x, y) and would like to capture the relationship between x and y. More formally, our goal is to learn a function h: X→Y so that given an unseen observation x, h(x) can confidently predict the corresponding output y.

In this module we will explore the inner workings of KNN, choosing the optimal K values and using KNN from scikit-learn.

## Split the dataset into training and test sets


### Question 7

Split the dataset into training and test sets with 80-20 ratio

Hint: use **train_test_split()**

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
# drop id column
iris_new = iris.drop(labels = "Id", axis = 1)

#i=only feature dataframe
iris_feature = iris_new.drop(labels= "Species" , axis = 1)
```

```python
# take Species values into some variable...series
iris_labels = iris_new['Species']
```

```python
# copy dataframe into array
iris_arr = np.array(iris_feature)

#copy labels into array
labels_arr = np.array(iris_labels)
```

```python
# now lets divide data into training set
Iris_train, Iris_test, iris_train, iris_test = train_test_split(iris_arr, labels_arr, test_size=0.2, random_state=1,stratify=iris_new['Species'])
```

## Build Model

```python
#Lets build model with = 3
iris_model = KNeighborsClassifier(n_neighbors= 3)

# fit the traiing data
iris_model.fit(Iris_train, iris_train)

#predict values
iris_predict = iris_model.predict(Iris_test)

#score
print("Score when k=3 ",iris_model.score(Iris_test, iris_test))

#Lets build model with = 5
iris_model = KNeighborsClassifier(n_neighbors= 3)

# fit the traiing data
iris_model.fit(Iris_train, iris_train)

#predict values
iris_predict = iris_model.predict(Iris_test)

#score
print("Score when k=5 ",iris_model.score(Iris_test, iris_test))

#Lets build model with = 9
iris_model = KNeighborsClassifier(n_neighbors= 9)

# fit the traiing data
iris_model.fit(Iris_train, iris_train)

#predict values
iris_predict = iris_model.predict(Iris_test)

#score
print("Score when k=9 ",iris_model.score(Iris_test, iris_test))


# It is observed that for these 3 values accuracy is not changing
```

### Question 10
