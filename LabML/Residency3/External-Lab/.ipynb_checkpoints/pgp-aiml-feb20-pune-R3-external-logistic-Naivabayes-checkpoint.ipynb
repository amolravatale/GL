## Drop variables with less variance

*Find the variance of each numerical independent variable and drop whose variance is less than 1.*

#Lets discuss some aspects of variance for numerical and categorical data
- For numerical data : The notion of variance (average of the squared difference from the mean) makes sense for attributes which exhibit "Continuity" aspect in its property/parameter.
- For Categorical data : The parameters which exhibit "Descrete" aspects in its propert/parameter, the variance which is built for numerical data does not make sense. In categorical data it is important to get idea of how often observations differ rather than how much.

So based on above concept, lets cosider only those attributes which are really numerical in nature and whose variance is less than 1...for dropping them from the dataset

```python
studentDf.var()
```

```python
# First lets create back of data set
studentdf_backup = studentDf       

# drop numerial attributes which have variance < 1
studentDf = studentDf.drop(columns=['traveltime','studytime','failures','famrel','freetime','Dalc'])
```

## Question 5 - Convert the continuous values of grades into classes

*Consider the values in G1, G2 and G3 with >= 10 as pass(1) and < 10 as fail(0) and encode them into binary values. Print head of dataframe to check the values.*

#### Answer:

```python
def PassFailFun(marks) :
    if (marks >= 10) :
        return 1
    else :
        return 0
```

```python
studentDf['G1']= studentDf['G1'].apply(PassFailFun)
```

```python
studentDf['G1'].value_counts()
```

```python
studentDf['G2']= studentDf['G2'].apply(PassFailFun)
```

```python
studentDf['G3']= studentDf['G3'].apply(PassFailFun)
```

## Question 10 - Model Implementation and Testing the Accuracy

*Build a **LogisticRegression** classifier using **fit()** and **predict()** functions in sklearn. Print Accuracy and F1 score of Training and Test sets.*

#### Answer:

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(train_set,train_labels)
```

```python
print ("Score for training data set=" + str(model.score(train_set,train_labels)))
```

```python
print ("Score for test data set=" + str(model.score(test_set,test_labels)))
```

```python
# Confusion matrix for prediction using test set
test_pred = model.predict(test_set)
print(metrics.confusion_matrix(test_labels, test_pred))
```

```python
print(metrics.classification_report(test_labels, test_pred))
```

```python
#Confusion matrix for prediction using training set
train_pred = model.predict(train_set)
print(metrics.confusion_matrix(train_labels, train_pred))
```

```python
print(metrics.classification_report(train_labels, train_pred))
```
