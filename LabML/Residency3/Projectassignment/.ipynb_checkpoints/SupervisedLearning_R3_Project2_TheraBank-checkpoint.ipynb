```python
#Function to check if any column has null values
def CheckifNullExists(sr) :
    if (sr.isnull().sum() >0) :
        return ("Column" + " " + "'" + sr.name + "'" + " " + "has " + sr.isnull().sum() + " values")
    else :
        return ("Column" + " " + "'" + sr.name + "'" + " " + "has no null values")
```

```python
# Check if any column has null values
for (columnName, columnData) in bankdf.iteritems():
    print(CheckifNullExists (columnData))
```

Some analysis on which attributes should be used for buidling model
Above pairplot shows that attributes Income is good attribute for building model. Also CCAvg and Mortgage attributes are good attributes for considering model building.
Reason is these attributes, based on above graphs, show good classification of people taken personal loan and not taken loan.
The distribution of Age, Education attributes based on personal loan does overlap...as shown in above graph...so logically will not help much for predicting if person will take loan or not...if we want to minimize campaign budget
Also attributes Securities Account, CD Account, Online, Creditcard, Education, family, Zip code do not provide resonable classification as observed in graph.

So we will go with 3 attributes : Income, CCAvg and Mortgage attributes for model building

```python
#Create new Dataframe by removing all unwanted attributes
dfBank = bankdf_new.drop(['Age','Experience','ZIP Code','Family','Education','Securities Account','CD Account','Online','CreditCard'],axis=1)
```

```python
#Lets build Logistic regression model for "dfBank" data set
```

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dfBank.drop("Personal Loan", axis=1)
```

```python
# labels
y= dfBank.pop("Personal Loan") 
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
model = LogisticRegression()
model.fit(train_set, train_labels)
```

```python
model.score(train_set , train_labels) 
```

```python
model.score(test_set , test_labels)  
```

```python
test_pred = model.predict(test_set)
print(metrics.confusion_matrix(test_labels, test_pred))
```

```python
print(metrics.classification_report(test_labels, test_pred))
```

```python
#Lets build KNN model
dfBank_knn = bankdf_new.drop(['Age','Experience','ZIP Code','Family','Education','Securities Account','CD Account','Online','CreditCard'],axis=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = dfBank_knn.drop("Personal Loan", axis=1)
```

```python
# labels
y= dfBank_knn.pop("Personal Loan") 
```

```python
#Lets split the data
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)
```

```python
# call KNN algorithm to fit the model
NKK = KNeighborsClassifier(n_neighbors=5, weights='distance')
```

```python
NKK.fit(X_train,y_train)
```

```python
predictaed_labels = NKK.predict(X_test)
```

```python
NKK.score(X_test,y_test)
```

```python
print(metrics.confusion_matrix(y_test,predictaed_labels))
```

```python
print(metrics.classification_report(y_test, predictaed_labels))
```

```python
#Lets build NaiveBayes model
dfBank_nv = bankdf_new.drop(['Age','Experience','ZIP Code','Family','Education','Securities Account','CD Account','Online','CreditCard'],axis=1)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
X = dfBank_nv.drop("Personal Loan", axis=1)
```

```python
# labels
y= dfBank_nv.pop("Personal Loan") 
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
# Call NB Gaussian function
model = GaussianNB()
model.fit(train_set, train_labels)
```

```python
model.score(train_set , train_labels)  
```

```python
test_pred = model.predict(test_set)
```

```python
print(metrics.confusion_matrix(test_labels, test_pred))
```

```python
print(metrics.classification_report(test_labels, test_pred))
```
