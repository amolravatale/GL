```python
df1['Height'] = df1['Height(Inches)'].map(lambda inches : round(inches * 0.0833,2))
```

```python
def fun1(val) :
    if (val > 5.5 and val <10) :
        return 6
    elif (val <= 5.5) :
        return 5
    elif (val > 60) :   
        return 60
    elif (val <= 60) :
        return 50

df1['NewHeight'] = df1['Height'].map(fun1)
```

```python
df1['Weight']=df1['Weight(Pounds)'].map(lambda Pounds: round(Pounds* 0.453,2))
```

```python
df1['NewWeight'] = df1['Weight'].map(fun1)
```
