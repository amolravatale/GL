```python
df_iris_normal = pd.DataFrame(sc.fit_transform(df_iris))
```

```python
df_iris_normal
```

```python
df_iris_normal.columns = features
```

```python
f1 = df_iris_normal['SepalLengthCm'].values
f2 = df_iris_normal['SepalWidthCm'].values
f3 = df_iris_normal['PetalLengthCm'].values
f4 = df_iris_normal['PetalWidthCm'].values
```

```python
X = np.array(list(zip(f1,f2,f3,f4)))
```

```python
cluster_range = range(2,6) 
cluster_error =[]
cluster_sil_scores = []
custer_silhouette_values = []
sample_silhouette_values = []
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
for cluster_num in cluster_range :
    model = KMeans(n_clusters=cluster_num, n_init=5)
    model.fit(X)
    cluster_error.append(model.inertia_)
    cluster_sil_scores.append(silhouette_score(X, model.fit_predict(X)))    
    sample_silhouette_values.append(silhouette_samples(X, model.fit_predict(X)))
```

```python
sample_silhouette_values[3]
```

```python
df1 = pd.DataFrame({"ClusterError" : cluster_error, "ClusterNo" : cluster_range})
```

```python
df2 = pd.DataFrame({"SilhouetteScore" :cluster_sil_scores,"ClusterNo" : cluster_range})
```

```python
plt.figure(figsize=(12,6))
plt.plot(df2.ClusterNo,df2.SilhouetteScore, marker="o")
```

```python
plt.figure(figsize=(12,6))
plt.plot(df1.ClusterNo,df1.ClusterError, marker="o")
```
