```python
X = dfhr_normal.values
```

```python
cluster_num = range (25,50)
cluster_errors = []
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
for cluster_count in cluster_num :
    model = KMeans(n_clusters=cluster_count, n_init=10)
    model.fit(X)
    cluster_errors.append(model.inertia_)
```

```python
df1 = pd.DataFrame({"ClusterError": cluster_errors, "ClusterNum" : cluster_num})
```

```python
df1
```

```python
plt.figure(figsize=(12,6))
plt.plot(df1.ClusterNum,df1.ClusterError, marker="o")
```

```python
dfsub = pd.read_csv("/home/amol/GITRepo/GL/Practise/HR/HR.txt",sep="\t",
                   usecols= ['Attrition','Age', 'BusinessTravel', 
                           'DistanceFromHome', 'Education', 
                           'EnvironmentSatisfaction', 
                           'JobSatisfaction',
                           'MaritalStatus', 'MonthlyIncome', 
                           'OverTime', 'YearsInCurrentRole']
                   )
```

```python
def attrition (val) :
    if val == 'Yes' :
        return 1
    else :
        return 0
```

```python
dfsub['Attrition'] = dfsub['Attrition'].apply(attrition)
```

```python
dfsub
```

```python
sns.pairplot(dfsub, hue='Attrition', diag_kind='kde')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
dfsub_normal = pd.DataFrame(sc.fit_transform(dfsub))
```

```python
dfsub_normal.columns = dfsub.columns
```

```python
X = dfsub_normal.values
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_num = range(40,70)
cluster_err = []
```

```python
for cluster_count in cluster_num :
    model = KMeans(cluster_count, n_init=10)
    model.fit(X)
    cluster_err.append(model.inertia_)
```

```python
dfelbow = pd.DataFrame({"ClusterError" : cluster_err, "ClusterCount" : cluster_num})
```

```python
plt.figure(figsize=(12,6))
plt.plot(dfelbow.ClusterCount,dfelbow.ClusterError, marker="o")
```
